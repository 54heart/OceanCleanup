```python
# Everything numerical grouped by State

byState = df.groupby("State").sum()
byState
```

```python
# Statewise % of each item collected
# slice relevant columns
grouped = df.groupby("State").sum().iloc[:, 6:53]
grouped
```

```python
# calculate percentage
grouped.div(grouped["Total Items Collected"], axis = 0).style.format('{0:,.2%}')
```

```python
# adult & children percentage for each state
byState.iloc[:, 0:3].div(byState["People"], axis = 0).style.format('{0:,.2%}')
```

```python
# Pounds per item for each state
byState["Pounds"].div(byState["Total Items Collected"], axis = 0)
```

```python
# Grouping trash into larger categories(8 categories in total)

df.loc[:,'Cigarette Materials'] = df['Cigarette Butts']+ df['Cigar Tips']+ df['Cigarette Lighters']+ df['Tobacco Packaging/Wrap']
df.loc[:, "Plastic Materials"] = df.loc[:, ['Take Out/Away Containers (Plastic)',  'Bottle Caps (Plastic)', 'Lids (Plastic)',  'Forks, Knives, Spoons',  'Straws, Stirrers',  'Beverage Bottles (Plastic)',  "Cups, Plates (Plastic)", "Other Plastic Bottles (oil, bleach, etc.)", "Strapping Bands"]].sum(axis = 1)
df.loc[:, "Foam Materials"] = df.loc[:, ['Take Out/Away Containers (Foam)',  'Cups, Plates (Foam)']].sum(axis=1)
df.loc[:, "Metal Materials"] = df.loc[:, ['Bottle Caps (Metal)',  'Beverage Cans']].sum(axis =1)
df.loc[:, "Glass Materials"] = df.loc[:, ['Beverage Bottles (Glass)']].sum(axis =1)
df.loc[:, "Paper Materials"] = df.loc[:, ['Paper Bags', 'Cups, Plates (Paper)', '6-Pack Holders']].sum(axis =1)
df.loc[:, "Others"] = df.loc[:, ['Balloons', 'Fireworks', 'Tires', 'Toys', 'Other Trash (Clean Swell)', '6-Pack Holders', 'Other Plastic/Foam Packaging', 'Other Plastic/Foam Packaging', 'Other Packaging (Clean Swell)',  'Fishing Buoys, Pots & Traps', 'Fishing Net & Pieces', 'Fishing Line (1 yard/meter = 1 piece)', 'Rope (1 yard/meter = 1 piece)', 'Fishing Gear (Clean Swell)',  'Appliances (refrigerators, washers, etc.)', 'Construction Materials',  'Condoms', 'Diapers', 'Syringes', 'Tampons/Tampon Applicators', 'Personal Hygiene (Clean Swell)']].sum(axis =1)
df.loc[:, "Tiny Trash"] = df.loc[:, ['Foam Pieces', 'Glass Pieces', 'Plastic Pieces'
]].sum(axis =1)
df.loc[:, "Larger Trash"] = df.loc[:, ['Cigarette Materials', 'Plastic Materials', 'Foam Materials', 'Metal Materials', 'Glass Materials', 'Paper Materials', 'Others']].sum(axis =1)
```
